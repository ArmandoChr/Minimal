```python
def height_to_int(height):
    if isinstance(height,str):
        if height.find('cm') != -1:
            height = height.replace('cm','')
            return int(height)
        elif height.find("'") !=-1:
            height = height.split("\'")
            h = int(height[0])*(30.48)+int(height[1].replace('"',''))*(2.54)
            return int(h)
        else:
            return height
    
```

```python
df['Height'] = df['Height'].apply(height_to_int).astype(int)
```

```python
df['Height'].unique()
```

```python
def weight_to_int(weight):
    if isinstance(weight,str):
        if weight.find('kg') != -1:
            weight = weight.replace('kg','')
            return int(weight)
        elif weight.find('lbs') != -1:
            weight = weight.replace('lbs', '')
            weight = int(weight) * (0.453592)
            return int(weight)
    else:
        return int(weight)
```

```python
df['Weight'] = df['Weight'].apply(weight_to_int).astype(int)
```

```python
def hits_data_clean(hits):
    if isinstance(hits,str):
        if hits.find('K') !=-1:
            hits = hits.replace('K','')
            return int(float(hits)*1000)
        else:
            return int(hits)
    else:
        return hits

```

```python
df['Hits'] = df['Hits'].apply(hits_data_clean).fillna(0).astype(int)
```

```python
def currency_data_clean(currency):
    if isinstance(currency,str):
        if currency.find('K') != -1:
            return int(currency.replace('K','').replace('€',''))*1000
        elif currency.find('M') != -1:
            return int(float(currency.replace('M','').replace('€',''))*1000000)
    else:
        return currency
```

```python
df['Value'] = df['Value'].apply(currency_data_clean).fillna(0).astype(int)
df['Wage'] =df['Wage'].apply(currency_data_clean).fillna(0).astype(int)
df['Release Clause'] = df['Release Clause'].apply(currency_data_clean).fillna(0).astype(int)
```

```python
def star_data_clean(star):
    if isinstance(star,str):
        if star.find('★') != -1:
            star = star.replace('★','')
            return int(star)
    else:
        return star
```

```python
df['IR'] = df['IR'].apply(star_data_clean).fillna(0).astype(int)
```

```python
df['SM'] = df['SM'].apply(star_data_clean).fillna(0).astype(int)
```

```python
df['W/F'] = df['W/F'].apply(star_data_clean).fillna(0).astype(int)
```
